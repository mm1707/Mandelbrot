```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("Mandelbrot.csv")
data["time_s"] = data["time_ns"] / 10**9

plt.plot(data["pixels"], data["time_s"], marker='o')
plt.xscale('log', base=2)
plt.yscale('log', base=2)
plt.title("Time to generate Mandelbrot vs plot size")
plt.xlabel("Plot size (number of pixels)")
plt.ylabel("time (s)")
plt.show()
```
